## Task 4: Image to Pencil Sketch with Python

**The objective of this project is to create a pencil sketch of normal image file**

```python
import cv2
```

```python
from matplotlib import pyplot as plt
```

```python
import matplotlib.image as mpimg
```

```python
img = cv2.imread("dog.jpg")
```

```python
plt.imshow(img)
b,q,r = cv2.split(img)
img = cv2.merge((r,q,b))
```

```python
plt.title("Original Image")
plt.imshow(img)
```

```python
img = mpimg.imread('dog.jpg')
R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B
plt.imshow(imgGray, cmap='gray')
plt.title("GrayScale Image")
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(8,8))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.title('Original image', size=10)
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('final sketch', size=10)
final=cv2.cvtColor( final , cv2.COLOR_BGR2RGB)
plt.imshow( final )
plt.axis('off')
plt.show()
```
